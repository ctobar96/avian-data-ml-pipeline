```python
# Configurar pandas para mostrar todas las columnas
pd.set_option('display.max_columns', None)

# 1. Ruta del archivo
ruta_Archivo = r'D:\crist\OneDrive\Estudios\Proyectos Para GitHub\avian-data-ml-pipeline\data\raw\Produccion\PRODUCCION_ENERO_2026.csv'

# 2. Leer el archivo con manejo de errores
try:
    # Leemos la pestaña correcta
    df_Produccion = pd.read_csv(ruta_Archivo, encoding='latin-1')  # Especificamos la codificación para evitar problemas con caracteres especiales  
    print(f"✅ Archivo leído correctamente: {ruta_Archivo}")
    print(f"📊 Total de filas cargadas: {len(df_Produccion)}\n")
    
    # 3. Ver los resultados (solo se ejecuta si el paso anterior funcionó)
    print("--- Información del DataFrame ---")
    df_Produccion.info()

    print("\n--- Primeras filas del DataFrame ---")
    display(df_Produccion.head())  # Usamos display para que se vea como tabla en Jupyter

except Exception as e:
    # Usamos ruta_Archivo en vez de df_Produccion para evitar el NameError
    print(f"❌ Error al intentar leer el archivo en {ruta_Archivo}: {e}")
```

### Limpieza de nombres de columnas

```python
df_Produccion.columns = (
    df_Produccion.columns
    .str.strip()
    .str.lower()
    .str.replace(' ', '_')
    .str.replace('.', '', regex=False)
    .str.normalize('NFKD')
    .str.encode('ascii', errors='ignore')
    .str.decode('utf-8')
)

df_Produccion.info()
```

### Dejar solamente las columnas a utilizar o utiles

```python
# 1. Definir una lista con los nombres de las columnas que necesitams
columnas_utiles = [
    'efectiva',
    'lin_producto',
    'tipo_trans',
    'numero_articulo',
    'um',
    'cantidad',
    'lote/serie',
    'almacen',
    'ubicacion'
]

# 2. Aplicar el filtro al DataFrame
df_Produccion_filtrado = df_Produccion[columnas_utiles]

# 3. Validar resultado
print(f"✅ Reducción exitosa: Pasamos de {len(df_Produccion.columns)} columnas a solo {len(df_Produccion_filtrado.columns)}.")
print("--- Nueva estructura del DataFrame ---")
display(df_Produccion_filtrado.head())
```

## 2. Preprocesamiento y Limpieza de Datos
En esta sección se realizan las transformaciones necesarias para asegurar la integridad de los datos, incluyendo el filtrado de transacciones relevantes, el tratamiento de valores nulos en el Lote/Serie y la normalización de cantidades.

### 2.1. Filtrado de Movimientos: Consumo (ISS-WO) y Producción (RCT-WO)
Filtramos el dataset para quedarnos únicamente con las transacciones de salida y entrada de producto terminado, eliminando ruidos operativos del ERP.

```python
# Definir las transformaciones de interes
transacciones_interes = ['RCT-WO', 'ISS-WO']

# Filtrar el DataFrame para quedarnos solo con las transacciones de interes
df_Produccion_filtrado = df_Produccion_filtrado[df_Produccion_filtrado['tipo_trans'].isin(transacciones_interes)]

# Validación de los datos filtrados
print(f"✅ Filtrado completado.")
print(f"Registro originales: {len(df_Produccion)} | Registros después del filtro: {len(df_Produccion_filtrado)}")
print(f"\nDistribución de las transacciones filtradas:")
print(df_Produccion_filtrado['tipo_trans'].value_counts())

# mostrar las primeras filas del DataFrame filtrado
df_Produccion_filtrado.info()
display(df_Produccion_filtrado.head())
```
