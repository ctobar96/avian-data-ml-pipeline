## 2. Preprocesamiento y Limpieza de Datos
En esta sección se realizan las transformaciones necesarias para asegurar la integridad de los datos, incluyendo el filtrado de transacciones relevantes, el tratamiento de valores nulos en el Lote/Serie y la normalización de cantidades.

### 2.1. Filtrado de Movimientos: Consumo (ISS-WO) y Producción (RCT-WO)
Filtramos el dataset para quedarnos únicamente con las transacciones de salida de ingredientes y entrada de producto terminado, eliminando ruidos operativos del ERP.

```python
# Definir los tipos de transacción que nos interesan
transacciones_interes = ['RCT-WO', 'ISS-WO']

# Filtrar el DataFrame para quedarnos solo con las filas que tengan los tipos de transacción deseados
df_alimento_filtrado = df_alimento_limpio[df_alimento_limpio['Tipo Trans'].isin(transacciones_interes)].copy()

# Validación de los datos filtrados
print(f"✅ Filtrado completado.")
print(f"Registros originales: {len(df_alimento_limpio)}")
print(f"Registros tras el filtro: {len(df_alimento_filtrado)}")
print(f"\nDistribución de transacciones en el nuevo set:")
print(df_alimento_filtrado['Tipo Trans'].value_counts())

# Mostrar los primeros resultados
display(df_alimento_filtrado.head())
df_alimento_filtrado.info()
```

### 2.2. Imputación de Valores Nulos (Forward Fill)
Los reportes del ERP agrupan visualmente los ingredientes omitiendo el nombre del lote en las filas secundarias. Aplicamos un "Forward Fill" para propagar el identificador del lote hacia abajo y asignar correctamente cada ingrediente a su orden de fabricación correspondiente.

```python
# Comprobación inicial: ¿Cuantos valores nulos tenemos?
nulos_antes = df_alimento_filtrado['Lote/Serie'].isna().sum()
print(f"🔍 Valores nulos en 'Lote/Serie' antes de la limpieza: {nulos_antes}")

# Aplicar Forward Fill
# Sobrescribimos la columna con su versión rellenada hacia abajo
df_alimento_filtrado['Lote/Serie'] = df_alimento_filtrado['Lote/Serie'].ffill()

# Comprobación final
nulos_despues = df_alimento_filtrado['Lote/Serie'].isna().sum()
print(f"✅ Valores nulos en 'Lote/Serie' DESPUÉS de limpiar: {nulos_despues}")

# 4. Ver visualmente cómo quedó rellenado
display(df_alimento_filtrado.head(10))
```

## 3. Transformación de Datos y Feature Engineering

En esta etapa, transformaremos los datos limpios en un formato numérico y estructurado que los algoritmos de Machine Learning puedan procesar eficientemente, además de crear nuevas variables que aporten mayor poder predictivo.

### 3.1 Extracción de Características Temporales (Feature Engineering)

Dado que la columna de fecha (`efectiva`) ya se encuentra nativamente en formato `datetime64[ns]`, procedemos directamente a extraer sus componentes numéricos. Esto permite que el modelo de Machine Learning identifique patrones estacionales, ciclos de producción semanales o tendencias anuales.

```python
# Extraer componentes cronológicos básicos
df_alimento_filtrado['AÑO'] = df_alimento_filtrado['Efectiva'].dt.year
df_alimento_filtrado['MES'] = df_alimento_filtrado['Efectiva'].dt.month
df_alimento_filtrado['DIA_DEL_MES'] = df_alimento_filtrado['Efectiva'].dt.day

# Extraer ciclos de producción
df_alimento_filtrado['DIA_DE_LA_SEMANA'] = df_alimento_filtrado['Efectiva'].dt.dayofweek # Lunes=0, Domingo=6
df_alimento_filtrado['SEMANA_DEL_AÑO'] = df_alimento_filtrado['Efectiva'].dt.isocalendar().week.astype(int)

# imprimir el resultado para validar
display(df_alimento_filtrado.head())
```

### 3.2 Mapeo binario para Tipo de Transacción
Asumiendo que RCT-WO es recibo/positivo y ISS-WO es emisión/consumo.
Para
- **RCT-WO: 1**
- **ISS-WO: 0**

```python
df_alimento_filtrado['Tipo Trans']= df_alimento_filtrado['Tipo Trans'].map({'RCT-WO': 1, 'ISS-WO': 0})

# Mostrar el resultado final para validar
display(df_alimento_filtrado.head())
```

### 3.2 Transformación y Creación de Nuevas Variables 
Nota: Si da > 0, gastaste de más. Si da < 0, ahorraste o pusiste de menos.

```python
# Convertir la cantidad real a positivo (valor absoluto)
df_alimento_filtrado['Cantidad'] = df_alimento_filtrado['Cantidad'].abs()

# Crear variable de diferencia (Varianza en kilos)
df_alimento_filtrado['Desviacion_Kg'] = df_alimento_filtrado['Cantidad'] - df_alimento_filtrado['Cnt Requerd']

# Crear la variable de Eficiencia (Ratio)
df_alimento_filtrado['Ratio'] = np.where(df_alimento_filtrado['Cnt Requerd'] == 0, 0, 
                                         df_alimento_filtrado['Cantidad'] / df_alimento_filtrado['Cnt Requerd'])

# Eliminar la variable Cnt Requerd, debido a puede presentar multicolonielidad de los datos
df_alimento_filtrado = df_alimento_filtrado.drop(columns=['Cnt Requerd'])

# Mostrar el resultado final para validar
display(df_alimento_filtrado.head(20))
```

### 3.3 Detección de Cambios de Fórmula (Versionado Dinámico)

Dado que la empresa recicla el identificador del lote (`Lote/Serie`) para distintas formulaciones a lo largo del tiempo, necesitamos crear un "Versionador de Recetas" para que el modelo sepa cuándo los cambios nutricionales son intencionales.

```python
# Calcular el tamaño del "batch" total ese día
df_alimento_filtrado['Batch_Diario'] = df_alimento_filtrado.groupby(['Efectiva', 'Lote/Serie'])['Cantidad'].transform('sum')

# Calcular el Porcentaje de la fórmula (Concentración de cada ingrediente en el batch)
# Esto nos da el peso relativo del ingrediente, ignorando si hicieron 1500 o 3000 kg
df_alimento_filtrado['Porcentaje_Formula'] = (df_alimento_filtrado['Cantidad'] / df_alimento_filtrado['Batch_Diario'])*100

# Redondear el porcentaje a 2 decimales para mejor legibilidad
df_alimento_filtrado['Porcentaje_Formula'] = df_alimento_filtrado['Porcentaje_Formula'].round(2).astype(float)

# Detectar cambio real de la formula
df_alimento_filtrado.sort_values(by=['Efectiva', 'Lote/Serie', 'Numero articulo'], inplace=True)  # Aseguramos que los datos estén ordenados cronológicamente

# 1. Calculamos si hubo un cambio (comparando con el anterior)
cambio_calculado = (
    df_alimento_filtrado.groupby(['Lote/Serie', 'Numero articulo'])['Porcentaje_Formula']
    .shift(1)
    .ne(df_alimento_filtrado['Porcentaje_Formula'])
)

# 2. Evitamos que el primer registro de un lote se marque como "cambio" (porque no hay anterior)
es_primer_registro = df_alimento_filtrado.groupby(['Lote/Serie', 'Numero articulo'])['Porcentaje_Formula'].shift(1).isna()
cambio_real = cambio_calculado & ~es_primer_registro # Es cambio SOLO si son diferentes Y NO es el primer registro

# np.where(condición, valor_si_verdadero, valor_si_falso)
df_alimento_filtrado['Cambio_Formula'] = np.where(
    df_alimento_filtrado['Tipo Trans'] == 0, 
    cambio_real, 
    False 
)

# 4. Crear la Versión Dinámica
df_alimento_filtrado['Version_Formula_Ingrediente'] = df_alimento_filtrado.groupby(['Lote/Serie', 'Numero articulo'])['Cambio_Formula'].cumsum() + 1
```
