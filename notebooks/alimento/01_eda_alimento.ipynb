## 2. Preprocesamiento y Limpieza de Datos
En esta sección se realizan las transformaciones necesarias para asegurar la integridad de los datos, incluyendo el filtrado de transacciones relevantes, el tratamiento de valores nulos en el Lote/Serie y la normalización de cantidades.

### 2.1. Filtrado de Movimientos: Consumo (ISS-WO) y Producción (RCT-WO)
Filtramos el dataset para quedarnos únicamente con las transacciones de salida de ingredientes y entrada de producto terminado, eliminando ruidos operativos del ERP.

```python
# Definir los tipos de transacción que nos interesan
transacciones_interes = ['RCT-WO', 'ISS-WO']

# Filtrar el DataFrame para quedarnos solo con las filas que tengan los tipos de transacción deseados
df_alimento_filtrado = df_alimento_limpio[df_alimento_limpio['Tipo Trans'].isin(transacciones_interes)].copy()

# Validación de los datos filtrados
print(f"✅ Filtrado completado.")
print(f"Registros originales: {len(df_alimento_limpio)}")
print(f"Registros tras el filtro: {len(df_alimento_filtrado)}")
print(f"\nDistribución de transacciones en el nuevo set:")
print(df_alimento_filtrado['Tipo Trans'].value_counts())

# Mostrar los primeros resultados
display(df_alimento_filtrado.head())
df_alimento_filtrado.info()
```

### 2.2. Imputación de Valores Nulos (Forward Fill)
Los reportes del ERP agrupan visualmente los ingredientes omitiendo el nombre del lote en las filas secundarias. Aplicamos un "Forward Fill" para propagar el identificador del lote hacia abajo y asignar correctamente cada ingrediente a su orden de fabricación correspondiente.

```python
# Comprobación inicial: ¿Cuantos valores nulos tenemos?
nulos_antes = df_alimento_filtrado['Lote/Serie'].isna().sum()
print(f"🔍 Valores nulos en 'Lote/Serie' antes de la limpieza: {nulos_antes}")

# Aplicar Forward Fill
# Sobrescribimos la columna con su versión rellenada hacia abajo
df_alimento_filtrado['Lote/Serie'] = df_alimento_filtrado['Lote/Serie'].ffill()

# Comprobación final
nulos_despues = df_alimento_filtrado['Lote/Serie'].isna().sum()
print(f"✅ Valores nulos en 'Lote/Serie' DESPUÉS de limpiar: {nulos_despues}")

# 4. Ver visualmente cómo quedó rellenado
display(df_alimento_filtrado.head(10))
```
